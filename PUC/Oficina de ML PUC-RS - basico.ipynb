```
import sklearn
```

```
from sklearn import datasets, tree
```

```
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
```

```
breast = load_breast_cancer()
```

```
 print (breast.DESCR)
```

```
breast.data[:5]
```

```
breast.target[:5]
```

```
breast.target_names
```

```
classifier = DecisionTreeClassifier()
classifier
```

```
classifier.fit(breast.data, breast.target)
```

```
classifier.predict(breast.data[:5])
```

```
from sklearn.neighbors import KNeighborsClassifier
clf_knn = KNeighborsClassifier()
clf_knn.fit (breast.data, breast.target)
clf_knn
```

```
len(breast.data)
```

```
clf_knn.fit(breast.data[:500], breast.target[:500])
clf_knn.predict(breast.data[500:])
```

```
breast.target[500:]
```

```
from sklearn.model_selection import cross_val_score
cross_val_score(clf_knn, breast.data, breast.target, cv=5)
```
