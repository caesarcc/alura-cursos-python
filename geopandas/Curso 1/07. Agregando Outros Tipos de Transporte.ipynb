# Estações de Trem

```python
#carrega dados de localização das estações de trem
trem = gpd.read_file('../Dados/Transporte/Trem/Estações_Trem.geojson')
```

```python
trem.head()
```

```python
trem.crs
```

```python
#converte o sistema de coordenadas para UTM e seleciona unidade KM
trem = trem.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
trem.crs
```

```python
#plota as informações sobre a base que é o mapa
base = rj.plot(color='white', edgecolor='black', figsize=(15, 8))
geo_dados.plot(ax=base, color='orange', alpha=0.2)
#adiciona localizações dos metrôs
metro.plot(ax=base, color='black', markersize=80)
#adiciona localização das estações de trem
trem.plot(ax=base, color='red', markersize=80)
#Há estaões de trem fora do RJ, que estão corretas, mas devem ser removidas pra esta pesquisa
```

```python
#limita os dados de localização de trens aos que estiverem dentro da geometria da cidade do RJ
trem = trem[trem.within(rj.iloc[0].geometry)]
```

```python
#plota novamente conforme a celula de plotagem anterior e agora somente com as localizações de trem dentro da cidade.
base = rj.plot(color='white', edgecolor='black', figsize=(15, 8))
geo_dados.plot(ax=base, color='orange', alpha=0.2)
metro.plot(ax=base, color='black', markersize=80)
trem.plot(ax=base, color='red', markersize=80)
```

# Estações BRT

```python
#carrega localização das estações BRT
brt = gpd.read_file('../Dados/Transporte/BRT/Estações_BRT.geojson')
```

```python
brt.head()
```

```python
brt.crs
```

```python
#converte o sistema de coordenadas para UTM e seleciona unidade KM
brt = brt.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
brt.crs
```

```python
#plota novamente os mesmos dados de antes adicionando agora o BRT.
base = rj.plot(color='white', edgecolor='black', figsize=(15, 8))
geo_dados.plot(ax=base, color='orange', alpha=0.2)
metro.plot(ax=base, color='black', markersize=80)
trem.plot(ax=base, color='red', markersize=80)
brt.plot(ax=base, color='blue', markersize=80)
```

```python
#concatena tods geometrias carregadas sobre o dataset de imóveis (trem, metro e brt)
# não mantei os índices para gerar novos índices para cada geometria adicionada (ignore_index=True)
#poderia identificar qual o tipo de transporte de cada linha de geometria, mas nao foi necessário para este exemplo
transporte = pd.concat([trem.geometry, metro.geometry, brt.geometry], ignore_index = True)
transporte[0:10]
```

```python
#cria coluna com a distância para cada imóvel até a geometria dos transportes disponíveis 
geo_dados['Dist_Transporte'] = geo_dados['geometry'].apply(lambda x: transporte.distance(x).min())
geo_dados.head()
```
