# Juntando os Dados Censitários ao Dataset

```python
%matplotlib inline
import geopandas as gpd
```

```python
#carregando dados do setor censitário mapeado no passo 02
setor = gpd.read_file('../Dados/Mapas/RJ-SETOR/RJ-SETOR-CENSO.shp', encoding="ISO-8859-1")
#carregando os dados do dataset de imóveis mapeado
geo_dados = gpd.read_file('../Dados/Mapas/RJ-DATASET/RJ-DATASET-FINAL.shp', encoding="ISO-8859-1")
```

```python
setor
```

```python
geo_dados
```

```python
#plota o setor censitário junto aos imóveis
base = setor.plot(color='white', edgecolor='gray', figsize=(15, 8))
geo_dados.plot(ax = base, color='blue', markersize=1)
```

```python
geo_dados.shape
```

```python
#juntar ao geo dados, os dados do setor censitário, sjoin é um JOIN ESPACIAL, que utiliza a geometria
#a operação de within vai testar se o ponto do setor está dentro do polígono dos imóveis (ou vice-versa)
geo_dados = gpd.sjoin(geo_dados, setor, how='left', op='within')
```

```python
geo_dados.shape
```

```python
#listas registros NÃO duplicados na junção como true
~geo_dados.index.duplicated(keep='first')
```

```python
#remove os registros duplicados do dataframe resultante da junção conforme a listagem anterior
geo_dados = geo_dados[~geo_dados.index.duplicated(keep='first')]
```
