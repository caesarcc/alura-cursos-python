# Criando Shapefile

## Lendo

```python
%matplotlib inline
import geopandas as gpd
```

```python
rs = gpd.read_file('Mapas/RS/43MUE250GC_SIR.shp', encoding='utf-8')
type(rs)
```

```python
rs.plot(color='white', edgecolor='black', figsize=(20,9))
```

```python
poa = rs[rs['NM_MUNICIP'] == 'PORTO ALEGRE']
poa.head(10)
```

```python
poa.plot(color='orange', edgecolor='black', figsize=(20,9))
```

## Salvando

```python
import os
mapa_poa = 'Mapas/RS-MUNIC'
if not os.path.exists(mapa_poa):
    os.makedirs(mapa_poa)
poa.to_file(mapa_poa + '/POA-MUNIC.shp')
```

```python
poa = gpd.read_file(mapa_poa + '/POA-MUNIC.shp', encoding='utf-8')
poa.plot(color='orange', edgecolor='black', figsize=(20,9))
```
