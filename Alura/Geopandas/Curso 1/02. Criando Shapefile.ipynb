# Criando Shapefile

## Lendo

```python
%matplotlib inline
import geopandas as gpd
```

```python
rj = gpd.read_file('../Dados/Mapas/RJ/33MUE250GC_SIR.shp', encoding='utf-8')
type(rj)
```

```python
rj.plot(color='white', edgecolor='black', figsize=(20,9))
```

```python
rj = rj[rj['NM_MUNICIP'] == 'RIO DE JANEIRO']
rj.head(10)
```

```python
rj.plot(color='orange', edgecolor='black', figsize=(20,9))
```

## Salvando

```python
import os
mapa_rj = '../Dados/Mapas/RJ-MUNIC'
if not os.path.exists(mapa_rj):
    os.makedirs(mapa_rj)
rj.to_file(mapa_rj + '/RJ-MUNIC.shp')
```

```python
rj = gpd.read_file(mapa_rj + '/RJ-MUNIC.shp', encoding='utf-8')
rj.plot(color='orange', edgecolor='black', figsize=(20,9))
```
