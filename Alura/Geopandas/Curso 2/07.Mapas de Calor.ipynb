```python
%matplotlib inline
import geopandas as gpd
import folium
```

```python
rj = gpd.read_file('../Dados/Mapas/RJ-MUNIC/RJ-MUNIC.shp')
setor = gpd.read_file('../Dados/Mapas/RJ-SETOR/RJ-SETOR-CENSO.shp')
```

```python
crs = {'init': 'epsg:4326'}
rj.to_crs(crs, inplace=True)
setor.to_crs(crs, inplace=True)
```

```python
import numpy as np
setor = setor.fillna(np.nan)
```

```python
#determina centro do mapa
y = rj.centroid.y.iloc[0]
x = rj.centroid.x.iloc[0]
```

```python
#gedando lista de pontos centrais de cada setor + a quantidade populacional de cada setor pelo valor m√°ximo que existe na coluna setor
data = []
for i in range(len(setor)):
    data.append([setor['geometry'][i].centroid.y,
                 setor['geometry'][i].centroid.x,
                 setor['V002'][i]/setor['V002'].max()])
```

```python
#remove valores nulos da lista
data = [x for x in data if ~np.isnan(x[2])]
data[:10]
```

```python
#importar plugin de mapa de calor do Folium
from folium.plugins import HeatMap
```

```python
base = folium.Map([y, x], zoom_start=11, tiles='OpenStreetMap')

HeatMap(data, name="Popula√ß√£o").add_to(base)

folium.LayerControl().add_to(base)

base.save('Folium Mapa de Calor.html')

base
```
