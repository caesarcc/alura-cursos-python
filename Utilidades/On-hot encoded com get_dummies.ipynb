```
import pandas as pd
import numpy as np

matriz = np.array([[1, 'F', 3], [4, 'M', 6], [4, np.nan, 6], [4, 'F', 6], [4, np.nan, 6], [4, 'M', 6]])
nomes_cols = ['C1', 'C2', 'C3']

df = pd.DataFrame(matriz, columns=nomes_cols)
pd.get_dummies(df, columns=['C2'])
```
