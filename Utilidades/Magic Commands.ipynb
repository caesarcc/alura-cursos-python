```
import numpy as np
from numpy.random import randint

# A function to simulate one million dice throws.
def one_million_dice():
    return randint(low=1, high=7, size=1000000)

# Let's try %time first
%time throws = one_million_dice()
%time mean = np.mean(throws)
```

```
# Let's do the same with %timeit
%timeit throws = one_million_dice()
%timeit mean = np.mean(throws)
```

```
# And finally %%time
%%time
throws = one_million_dice()
mean = np.mean(throws)
```

```
# "timeit" para medir tempo de execução de comandos
import timeit
#medir a execução de cada comando 10 mil vezes
print(timeit.timeit('"-".join(str(n) for n in range(100))', number=10000))
print(timeit.timeit('"-".join([str(n) for n in range(100)])', number=10000))
print(timeit.timeit('"-".join(map(str, range(100)))', number=10000))
# assim é possível identificar qual a forma mais rápida de execução em quantidades grandes
```

```
# testando tempo com magic command
import numpy as np
from numpy.random import randint
# funcao para simular um milhão de valores gerados.
def one_million_dice():
    return randint(low=1, high=7, size=1000000)
%time throws = one_million_dice()
%time mean = np.mean(throws)

%timeit throws = one_million_dice()
%timeit mean = np.mean(throws)
```
